```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
IMG_PATH = 'dog2.jpg'
```

```python
image = cv2.imread(IMG_PATH, cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
plt.hist(image.ravel(), bins = 256, range = [0,256])
plt.title("Histogram of Original Image")
plt.show()
```

---
### Contrast Streching

#### output = (image - min) * (255 / (max - min))
---

```python
min_val = np.min(image)
max_val = np.max(image)
```

```python
contrast_stretched = (
    (image - min_val) *
    (255 / (max_val - min_val))
).astype(np.uint8)
```

```python
plt.imshow(contrast_stretched, cmap='gray')
plt.title("Contrast Stretched Image")
plt.axis('off')
plt.show()
```

```python
plt.hist(contrast_stretched.ravel(), bins = 256, range = [0,256])
plt.title("Histogram of Contrast Stretched Image")
plt.show()
```

```python
kernel = np.ones((3,3), np.float32)/9
linear_filtered = cv2.filter2D(contrast_stretched, -1, kernel)
```

```python
plt.imshow(contrast_stretched, cmap='gray')
plt.title("Linear Filtered Image")
plt.axis('off')
plt.show()
```

```python
plt.hist(linear_filtered.ravel(), bins = 256, range = [0,256])
plt.title("Histogram of Linear Filtered Image")
plt.show()
```

```python
plt.figure(figsize=(10,20))
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(contrast_stretched, cmap='gray')
plt.title("Contrast Stretched Image")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(linear_filtered, cmap='gray')
plt.title("Linear Filtered Image")
plt.axis('off')

plt.show()
```

```python
plt.figure(figsize=(20,3))

plt.subplot(1, 3, 1)
plt.hist(image.ravel(), bins = 256, range = [0,256])
plt.title("Original Image")

plt.subplot(1, 3, 2)
plt.hist(contrast_stretched.ravel(), bins = 256, range = [0,256])
plt.title("Stretched Image")

plt.subplot(1, 3, 3)
plt.hist(linear_filtered.ravel(), bins = 256, range = [0,256])
plt.title("Linear Filtered Image")
plt.show()
```
